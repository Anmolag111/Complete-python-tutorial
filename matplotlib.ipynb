```python
x = np.arange(-10, 21)
y = x ** 2
plt.plot(x, y, label='square')

x2 = np.arange(-10, 21)
y2 = 10*x2 + 30
plt.plot(x2, y2, 'c--', linewidth=4, label='linear') # styling is always the third value

plt.legend()
plt.grid()

plt.text(14, 190, 'Hello')

plt.show()
```

```python
x = np.random.randint(1, 10, 20)
y = np.random.randint(1, 10, 20)

plt.scatter(x+10, y+10)
plt.scatter(x, y, color='r')
plt.show()
```

```python
years = [2000, 2004, 2008, 2012]

android_a = [8, 6, 3, 1]
android_b = [2, 3, 5, 3]
android_c = [0, 1, 1, 6]

plt.plot([],[],label='Android A', color='r')
plt.plot([],[],label='Android B', color='g')
plt.plot([],[],label='Android C', color='b')

plt.stackplot(years, android_a, android_b, android_c, colors=['r', 'g', 'b'])

plt.legend()
plt.show()
```

```python
slicess = [10,8,6,4]
cities = ['Delhi', 'Noida', 'Gurgaon', 'Chandigarh']

plt.pie(slices, labels=cities, explode=(0,0,0,0.2), startangle=90)

plt.show()
```
